```python

# Columns for outlier treatment
columns_to_treat = ['room_bed', 'room_bath', 'basement', 'Latitude','Longitude']

# Function to perform outlier treatment using IQR
def treat_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace outliers with median value
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), df[column].median(), df[column])
    
# Perform outlier treatment for each column
for col in columns_to_treat:
    treat_outliers(df, col)
```

```python
from scipy import stats 
# Columns for outlier treatment
outlier_columns = ['living_measure', 'lot_measure', 'ceil_measure', 
                   'living_measure2015', 'lot_measure2015', 'total_area', 'age_built']

# Function to perform outlier treatment using z-score
def treat_outliers_zscore(df, column, threshold=3):
    z_scores = stats.zscore(df[column])
    outliers = (np.abs(z_scores) > threshold)
    df[column] = np.where(outliers, np.nan, df[column])

# Perform outlier treatment for each column
for col in outlier_columns:
    treat_outliers_zscore(df, col)
```

```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming df is your DataFrame and it's already loaded
data = pd.DataFrame(df)

# Handle missing values
data = data.dropna()

# Split the data into features and target variable
X = data.drop(columns=['price'])  # Assuming 'price' is the target variable
y = data['price']

# Instantiate DictVectorizer
vec = DictVectorizer()

# Fit and transform the data
X = vec.fit_transform(X.to_dict(orient='records'))

# Convert sparse matrix to DataFrame
X_df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names_out())

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=42)

# Instantiate LightGBM model
lgbm_model = LGBMRegressor()

# Train the model
lgbm_model.fit(X_train, y_train)

# Make predictions
lgbm_pred = lgbm_model.predict(X_test)

# Calculate evaluation metrics for the model
lgbm_mae = mean_absolute_error(y_test, lgbm_pred)
lgbm_mse = mean_squared_error(y_test, lgbm_pred)
lgbm_r2 = r2_score(y_test, lgbm_pred)

# Display the evaluation metrics for the model
print("LightGBM:")
print(f"MAE: {lgbm_mae}, MSE: {lgbm_mse}, R2: {lgbm_r2}")
```

```python
from sklearn.model_selection import cross_val_score

# Define the Random Forest model
lgbm_model = LGBMRegressor()

# Perform cross-validation
cv_scores = cross_val_score(lgbm_model, X, y, cv=5)  # cv=5 means 5-fold cross-validation

# Print the cross-validation scores
print("Cross-Validation Scores:", cv_scores)

# Print the mean and standard deviation of the cross-validation scores
print("Mean CV Score:", cv_scores.mean())
print("Standard Deviation of CV Scores:", cv_scores.std())
```

```python
import pickle

# Define the filename
filename = 'final_model.pkl'

# Load the model from disk
with open(filename, 'rb') as file:
    loaded_model = pickle.load(file)

# Fit the loaded model to your training data
loaded_model.fit(X_train, y_train)

# Make predictions using the loaded and fitted model
result = loaded_model.score(X_test, y_test)
print(f"{result * 100}% Accuracy")  # Print the accuracy of the loaded model

```
